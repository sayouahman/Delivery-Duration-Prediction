```python
#Ici on essaye de visualiser le dataset pour avoir une idée
import pandas as pd
histor_data =  pd.read_csv('historical_data.csv')
#print(histor_data.head())
```

```python
#On veut calculer delivery time 
print(histor_data['created_at'].dtype)
```

```python
histor_data['created_at'] = pd.to_datetime(histor_data['created_at'])
```

```python
print(histor_data['created_at'].isna().sum())
```

```python
histor_data['actual_delivery_time'] = pd.to_datetime(histor_data['actual_delivery_time'])
```

```python
#Transmettre delivery time to hours pour avoir une vision sur cette durée 
histor_data['delivery_hours'] = (histor_data['actual_delivery_time'] -  histor_data['created_at']).dt.total_seconds()/3600
histor_data = histor_data[histor_data['delivery_hours']<20]
```

```python
#Eliminer les outliers
Q1 = histor_data['delivery_hours'].quantile(0.25)
Q3 = histor_data['delivery_hours'].quantile(0.75)
IQR = Q3 - Q1
upperband = Q3 + 1.5*IQR
lowerband = Q1 - 1.5*IQR
histor_data = histor_data[(histor_data['delivery_hours']<= upperband) & (histor_data['delivery_hours']>=lowerband)]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.scatter(histor_data.index, histor_data['delivery_hours'])
plt.title('Delivery Duration per Order')
plt.xlabel('Order ID (row index)')
plt.ylabel('Delivery Duration (hours)')
plt.show()
```

```python
plt.hist(histor_data['delivery_hours'], bins=5)
plt.title('a')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Essayer de trouver les variables influancant cette durée et donc étuliser la matrice de correlation 
import seaborn as sns
corr_matr = histor_data.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr_matr, cmap='coolwarm')
plt.title("Correlation Matrix of All Numeric Features")
plt.show()
```

```python
#Prendre specifiquement la matr
d_corr = corr_matr['delivery_hours'].sort_values(ascending=False)
print(d_corr)
```
